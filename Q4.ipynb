#Discuss the different types of operators in Python and provide examples of how they are used


#Operators:
Operators are used to perform operations on values and variables. There are some symbols used for logical and arithmetic operations.

To perform any operations we need operators and operands:

A.Operators: Special symbols are to perform some

B. Operand: It is the value on which the operator is applied.

#Types of Operators:
1. Airthmatic Operators: These Operators is used to perform for basic mathematical operations i.e. addition,subtraction, multiplication and Division.
  
  A. "+, -, *, /, %"that are used to perform Arithmatic operations.

  B. "%" symbol is used to get remainder.it is also called modules.
2. Comparison Operators: These Operators are used to compares the values.

  A.It either returns True or False according to the condition
  
  B. "<, >, ==, !=, <=, >=" these all are Comparision symbols.

3. Logical Operators: It is used to perform Logical AND, Logical OR, and Logical NOT Operation.

  A. It is used to combine conditional statements.

  B."and , or , not" are the keywords which is used to perform the Logical Operations.
4. Assignment Operators: It is used to assign values to the variables.

  A. "=, +=, -=, *= , /= ,etc" are used the assignment operators.

5. Bitwise Operators: It acts on bits and perform bit-by-bit operations.

  A. "&, |, ~, ^" are the Bitwise Opertions.
  
  B.These are used to operate on binary numbers.

```python
# performing Arithmatic Operation and Assignment Operation

# Taking Input from the user

num1 = int(input("Enter First Number : "))
num2 = int(input("Enter Second Number : "))

# Performing Additon, subtraction, multiplicaiton, division and modulas
print(f"Addition of two numbers are {num1 + num2}")
print(f"Subtraction of two numbers are {num1 - num2}")
print(f"Multiplication of two numbers are {num1 * num2}")
print(f"Division of two numbers are {num1 / num2}")
print(f"Modulus of two numbers are {num1 % num2}")
```

```python
# Comparision Operation using if else

name = ['Aman','Ajay','Suman','Ayush']      #list of Names

name1 = input("Enter Your Name : ")       #Taking name as a Input

for i in name:
  if (name1 == i):      #Comparing input with the list
    print(f"{name1} is present in the list")
    break
else:
  print(f"{name1} is not present in the list")
```
