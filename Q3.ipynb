#Compare and contrast mutable and immutable objects in python with examples.

###Mutable:
1. It allows us to change their value or data in place without affecting their identity.
2. We can modify their internal state after creation
3. List and dictionary are the examples of mutability.

###Immutable:
1. It doesn't allows us to changes their state once they are created.
2. You can create new objects of the same types of the same types with different values.
3. Strings and Tuples are the examples of Immutability.
