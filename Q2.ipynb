#Describe the role of predefined keywords in Python and provide examples of how they are used in a program.

##Role of Predefined keywords in python:
1. Predefined keywords are reserved words that can not be used as a variable or function or any other identifier.
2. predefined keywords having some specific task they can only uesd for that task only.
3. In python there are 35 keyword like for,if,else,elif etc
4. All the keywords in Python should be in lowercase except True and False.

```python
#Checking Entered string is vowel or not.

Str = input("Enter a character: ") #storing the input in str variable

if(Str == 'a' or Str == 'e' or Str == 'i' or Str == 'o' or Str == 'u' or Str == 'A' or Str == 'E' or Str == 'I' or Str == 'O' or Str == 'U'):
  print("Entered character is vowel")
else:
  print("Entered character is not vowel")
```

```python
#in this Example for, in, range and print keywords are used

#printing a square using * (asterisk)

for i in range(5):          #for(keyword) loop for rows
  for j in range(5):
    print("*",end = " ")
  print()
```
