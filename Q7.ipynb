#Describe the different types of loops in Python and their use cases with examples.

####Python provides two types of loops to handle looping:
  1. For loop: It is used for iterating over a sequence (that is either a list, a tuple,).
  
    A. for loop we can execute a set of statements, once for each item in a list, tuple, set etc.

    B. 'for' keyword is used for using for loop

2. While loop: we can execute a set of statements as long as a condition is true.


```python
#print the table of 2

for i in range(0,21,2):       #iterates every time from the range
  if(i%2==0 and i!=0):        #check if the number is divisible by 2 and i is not equal to 0
    print(i)
```

```python
languages = ['Hindi','English','Bhojpuri','Marathi']    #list of languages

for i in languages:             #for loop is visiting all the elements of the list and priting in the sequence
  print(i)
```

```python
i = 1
while i < 6:
  print(i)
  i += 1
```

```python
#printing the 3*3 square matrix by using for loop
for i in range(3):
  for j in range(3):
    print("*", end = " ")
  print()
```
