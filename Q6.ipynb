#How do conditional statements work in Python? Illustrate with Examples.

####Conditional Statements are statements in Python that provide a choice for the control flow based on a condition.

####Types of Conditional Statement:
There are three Conditional Statement
1. if Statement
2. if-else Statement
3. if-elif-else statement

1.if statement: It is a conditional statement in Python that is used to execute a block of code only when a specific condition is met.

    A. If the Condition is True, the body of the if statement is executed.
    B. If the Condition is False, the body of the if statement is skipped.
    
2.
 if-else statement: It is a conditional statement in Python that is used to execute block of code only when a specific condition is met otherwise execute else block.

    A. If the Condition is True, the body of the if statement is executed
    
    B. If the condition is False, the body of the else statement is executed
    

3.if-elif-else statement:  It is a conditional statement in Python that is used to execute block of code only when a specific condition is met otherwise execute elif block otherwise else block is execute.

    A. If the Condition is True, the body of the if statement is executed.
    B. If the Condition is False, the body of the elif statement is executed.
    C. If the Condition of both if or elif is False, then the body of the else statement is executed.
    D. We can Add multiple elif block in a single Code.

```python
# If condition

x = int(input("Enter a number : "))   #taking input a integer
if x > 5:                           #checking the number is greater than 5
  print("x is greater than 5")
print("if Block of code is running successfully")
```

```python
#if-else condition

x = int(input("Enter a number : "))      #taking input a integer
if x > 5:                              #Checking the number is greater or less than 5
  print("x is greater than 5")
else:
  print("x is less than 5")
print("if-else Block of code is running successfully")
```

```python
#if-elif-else Condition

x = int(input("Enter a number : "))      #taking input a integer
if x > 5:                              #Checking the number is greater or less than 5
  print(f"{x} is greater than 5")
elif x == 5:
  print(f"{x} is equal to 5")
else:
  print(f"{x} is less than 5")
print("if-elif-else Block of code is running successfully")
```
