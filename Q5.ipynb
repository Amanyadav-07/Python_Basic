#Expalin the concept of type casting in Python with examples.

###Type Casting:
It is the method to convert the Python variable datatype into a certain data type in order to perform the required operation by users

There can be two types of Type Casting in Python:

1. Implicit Type Conversion: Python converts the datatype into another datatype automatically.Users don't have to involve in this process.

2. Explicit Type Conversion: Python needs user inivolvement to convert the variable data types into the required data types.

    A. While Converting string like any name or character can not be converted in Float or Integer data types.

```python
# implicit conversion

name = 'aman'       #string data types
type(name)
```

```python
# implicit conversion

num1 = 50          #integer data types
type(num1)
```

```python
# implicit conversion

num2 = 50.5        #float data types
type(num2)
```

```python
# Explicit Conversion

a = '7'           #Conversion of string to integer
type(int(a))
```

```python
# Explicit conversion

f1 = 7.5        #conversion of float data types to integer
type(int(f1))
```

```python
# Explicit Conversion

b = 7.5        #conversion of float data types to string
type(str(b))
```

```python
# Explicit Conversion

c = 7          #conversion of integer data types to string
type(str(c))
```

```python
# Explicit Conversion

d = 7          #conversion of integer data types to float
type(float(d))
```

```python
# Explicit Conversion

name1 = 'aman'   #conversion of string data types to float
type(float(name1))
```

```python
# Explicit Conversion

name2 = 'aman'   #conversion of string data types to integer
type(int(name2))
```
